## REFERENCE SHEET

### sklearn.neighbors - KNeighborsRegressor
```python
from sklearn.neighbors import KNeighborsRegressor

# Create model
knn = KNeighborsRegressor(n_neighbors=5, weights='uniform', metric='minkowski', p=2)
# n_neighbors: Number of neighbors to use (k)
# weights: 'uniform' (all equal) or 'distance' (closer neighbors have more influence)
# metric: Distance metric ('euclidean', 'manhattan', 'minkowski')
# p: Power parameter for Minkowski metric (p=1 -> Manhattan, p=2 -> Euclidean)

# Train model
knn.fit(X_train, y_train)

# Make predictions
y_pred = knn.predict(X_test)

# Get R² score
score = knn.score(X_test, y_test)
```

### sklearn.metrics - Regression Metrics
```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Mean Squared Error (MSE)
mse = mean_squared_error(y_true, y_pred)

# Root Mean Squared Error (RMSE)
rmse = mean_squared_error(y_true, y_pred, squared=False)
# OR: rmse = np.sqrt(mean_squared_error(y_true, y_pred))

# Mean Absolute Error (MAE)
mae = mean_absolute_error(y_true, y_pred)

# R² Score (Coefficient of Determination)
r2 = r2_score(y_true, y_pred)
```

### Distance Metrics
```python
# Euclidean distance (L2 norm) - Default
knn = KNeighborsRegressor(n_neighbors=5, metric='euclidean')
# OR
knn = KNeighborsRegressor(n_neighbors=5, metric='minkowski', p=2)

# Manhattan distance (L1 norm)
knn = KNeighborsRegressor(n_neighbors=5, metric='manhattan')
# OR
knn = KNeighborsRegressor(n_neighbors=5, metric='minkowski', p=1)

# Chebyshev distance
knn = KNeighborsRegressor(n_neighbors=5, metric='chebyshev')
```

import matplotlib.pyplot as plt

- Small k: High variance (overfitting), Low bias
- Large k: Low variance, High bias (underfitting)
- Feature scaling is crucial for distance-based algorithms
- Different distance metrics can impact performance
