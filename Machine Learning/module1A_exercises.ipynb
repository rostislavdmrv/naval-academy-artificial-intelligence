**Какво прави кодът:**
- ДВЕ неща:

**Важна теория - fit vs transform:**
- **fit()**: "Учи" от данните (изчислява mean, std)
- **transform()**: Прилага научените параметри
- **fit_transform()**: fit + transform заедно (за train data)

**КРИТИЧНО ВАЖНО за train/test:**
- Train data: fit_transform() (учи И прилага)
- Test data: само transform() (прилага СЪЩИТЕ параметри от train)
- НИКОГА не правим fit на test data! (data leakage)

## Част 5: Train/Test Split

```python
# Задача 5.1: Импортирайте train_test_split
from sklearn._____ import _____
```

**Защо го правим:**
- Train/Test split е ФУНДАМЕНТАЛНА концепция в ML
- Разделяме данните за обучение и тестване

**Важна теория - Train/Test Split:**
- **Проблем**: Ако тестваме на същите данни, с които обучаваме - модела запомня (overfitting)
- **Решение**: Разделяме данните на две части
  - **Train set**: За обучение на модела (обикновено 70-80%)
  - **Test set**: За оценка на модела (обикновено 20-30%)
- Модела НИКОГА не вижда test set по време на обучение

- **X**: Всички колони ОСВЕН price (features, independent variables)
- **y**: Само price колоната (target, dependent variable)

**Защо го правим:**
- ML модели имат формата: y = f(X)
- X = входните данни (characteristics на къщата)
- y = изходът който искаме да предскажем (цената)

**Терминология:**
- **X**: Features, predictors, independent variables, input

## Част 9: Евалуация на модела

```python
# Задача 9.1: Изчислете Mean Squared Error (MSE)
mse = _____(y_test, y_pred)
print(f"Mean Squared Error (MSE): {mse:,.2f}")
```

**Какво прави кодът:**
- Изчислява MSE (Mean Squared Error)
- Формула: MSE = (1/n) * Σ(actual - predicted)²

**Защо MSE:**
- Наказва големите грешки повече (заради квадрата)
- Винаги положителна стойност
- По-малка MSE = по-добър модел

**Недостатък на MSE:**
- Единиците са квадратни (price² !?)
- Трудно за интерпретация
- Затова използваме RMSE (квадратен корен от MSE)

```python
# Задача 9.2: Изчислете Root Mean Squared Error (RMSE)
rmse = np._____(mse)
print(f"Root Mean Squared Error (RMSE): {rmse:,.2f}")
print(f"Интерпретация: Средната грешка в прогнозата е ±{rmse:,.0f} единици")
```

**Какво прави кодът:**
- RMSE = √MSE
- Връща грешката в оригиналните единици

**Защо RMSE:**
- ИНТЕРПРЕТИРУЕМ! Същите единици като target
- "В среден случай, грешим с ±RMSE"
- По-малка RMSE = по-добър модел

**Интерпретация:**
- RMSE = 1,200,000 означава: "Средно грешим с ±1.2 милиона"
- Сравнете с диапазона на price (1.75M - 13.3M)
- Ако RMSE е малка спрямо диапазона → добър модел


```python
# Задача 9.3: Изчислете Mean Absolute Error (MAE)
mae = _____(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae:,.2f}")
```

**Какво прави кодът:**
- MAE = (1/n) * Σ|actual - predicted|
- Средната абсолютна грешка

**Защо MAE:**
- По-устойчив към outliers от RMSE
- НЕ наказва големи грешки толкова силно
- Интуитивна интерпретация

**MAE vs RMSE:**
- MAE винаги ≤ RMSE
- Ако RMSE >> MAE → има outliers с големи грешки
- Ако RMSE ≈ MAE → грешките са равномерни


```python
# Задача 9.4: Изчислете R² Score
r2 = _____(y_test, y_pred)
print(f"R² Score: {r2:.4f}")
print(f"Интерпретация: Моделът обяснява {r2*100:.2f}% от вариацията в цените")
```

**Какво прави кодът:**
- R² (R-squared, coefficient of determination)
- Най-важната метрика за regression!

**Формула и интерпретация:**
- R² = 1 - (SS_res / SS_tot)
- SS_res = Σ(actual - predicted)² (residual sum of squares)
- SS_tot = Σ(actual - mean)² (total sum of squares)
- **R² показва: Каква част от вариацията е обяснена от модела**

**Диапазон на стойности:**
- **R² = 1.0**: Перфектен модел (100% обяснена вариация)
- **R² = 0.8**: Много добър модел (80% обяснена вариация)
- **R² = 0.5**: Среден модел (50% обяснена вариация)
- **R² = 0.0**: Моделът не е по-добър от средната стойност
- **R² < 0**: Моделът е по-лош от средната стойност (много лошо!)

**За Simple Linear Regression с area:**
- Защо не по-високо: Само 1 feature (area)
- Price зависи и от: location, bedrooms, condition, etc.
- С всички features → R² може да е 0.7-0.8+

**КРИТИЧНО ВАЖНО за преподаване:**
- R² НЕ Е "процент на точност"!
- R² е "процент на обяснена вариация"
- R² = 0.5 НЕ означава "50% от прогнозите са точни"
- R² = 0.5 означава "модела обяснява 50% от разликите в цените"

**Пример:**
- Защо една къща струва 8M, а друга 4M?
- R² = 0.5 означава: "50% от тази разлика се обяснява с area"
- Останалите 50% се обясняват с други фактори (не в модела)
