## REFERENCE SHEET: Classification Metrics

### Confusion Matrix
```
                Predicted
              Pos       Neg
Actual  Pos   TP        FN
        Neg   FP        TN
```

### Metrics Formulas
- **Accuracy** = (TP + TN) / (TP + TN + FP + FN)
- **Precision** = TP / (TP + FP) - "Of all predicted positive, how many are correct?"
- **Recall (Sensitivity)** = TP / (TP + FN) - "Of all actual positive, how many did we find?"
- **F1-Score** = 2 * (Precision * Recall) / (Precision + Recall)

### k-Nearest Neighbors (kNN)
- **Algorithm:** Find k closest training samples to a new point
- **Classification:** Majority vote among k neighbors
- **Distance:** Usually Euclidean distance
- **Hyperparameter k:** Number of neighbors to consider
